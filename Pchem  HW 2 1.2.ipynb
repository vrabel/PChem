```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
MW = 32e-3        # kg/mol
Na = 6.022e23 # molecules/mol
k = 1.381e-23 # m^2 kg/s^2/K
R = 8.3144    # J/mol/K

V = np.linspace(0,1500,10000)

def Pnorm(V,T):
    return MW*V/(R*T)*np.exp(-(MW*V**2/(2*R*T)))

plt.figure()
for T in [200, 400, 600]:
    Prob = Pnorm (V,T)
    plt.plot(V,Prob)
    plt.title('2-Dimensional Speed Distribution for $O_2$')
    plt.xlabel('Speed (m/s)')
    plt.ylabel('Probability')

plt.legend(['200 K','400 K','600 K'])
```
